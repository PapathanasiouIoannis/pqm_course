### Συνεκτικές (coherent) καταστάσεις του απλού αρμονικού ταλαντωτή (ΑΑΤ)

Τώρα που γνωρίζουμε την κατασκευή των stationary καταστάσεων, θα διερευνήσουμε πώς εξελίσσονται στο χρόνο.  Θα ξεκινήσουμε εξετάζοντας πώς εξελίσσονται οι αυθαίρετες υπερθέσεις, και μετά θα προχωρήσουμε στην έννοια των **συνεκτικών καταστάσεων** (coherent states).

Mια συνεκτική κατάσταση, $\psi^{(\alpha)}$, είναι μια ειδική υπέρθεση των ενεργειακών ιδιοσυναρτήσεων του ΑΑΤ που ελαχιστοποιούν την αρχή της αβεβαιότητας.  Ορίζονται ως *ιδιοσυναρτήσεις του τελεστή υποβίβασης*.
$$ \hat{a}_- \psi^{(\alpha)} = \alpha \psi^{(\alpha)}$$
Όπου $ \alpha $ μπορεί να είναι οποιοσδήποτε μιγαδικός αριθμός.

Οποιαδήποτε αυθαίρετη κατάσταση του AAT, $\psi$, μπορεί να εκφραστεί ως υπέρθεση των ενεργειακών ιδιοκαταστάσεων, $\psi_n$.
$$ \psi = \sum_{n = 0}^\infty c_n \psi_n$$

Μια συνεκτική κατάσταση, $\psi^{(\alpha)}$, είναι απλά μια κατάσταση υπέρθεσης με πολύ συγκεκριμένες τιμές του $c_n$.

$$\psi^{(\alpha)}=\sum_{n=0}^\infty c_n^{coherent}\psi_n$$
$$c_n^{coherent} = \frac{\alpha^n}{\sqrt{n!}}c_0^{coherent}$$
$$c_0^{coherent} = exp(-|\alpha|^2/2)$$


Αυτό που δεν είναι τόσο προφανές από τα μαθηματικά είναι η πιο ενδιαφέρουσα ιδιότητα των συνεκτικών καταστάσεων: είναι οι κβαντικές καταστάσεις που μοιάζουν περισσότερο με τα κλασικά σωματίδια.  Η απεικόνιση και η σύγκριση της χρονικής εξέλιξης των συνεκτικών καταστάσεων με αυθαίρετες καταστάσεις θα καταστήσει εύκολα αυτή τη διάκριση.  Αυτή η ιδιότητα καθιστά τις coherent states ένα χρήσιμο εργαλείο σε πολλούς διαφορετικούς τομείς της φυσικής, όπως η κβαντική οπτική και η θεωρία πεδίου συμπυκνωμένης ύλης.



#### Ενεργειακές ιδιοκαταστάσεις του ΑΑΤ
Οι κυματοσυναρτήσεις και οι ενέργειες για τον ΑΑΤ δίνονται από τις σχέσεις:
$$\psi_n(x) = \frac{1}{\sqrt{2^nn!}}\left( \frac{m\omega}{\pi \hbar} \right)^{1/4}e^{-\frac{m\omega x^2}{2\hbar}}H_n\left(\sqrt{ \frac{m \omega}{\hbar}}x \right)$$
$$E_n=\hbar \omega \left( n+\frac{1}{2}\right) $$

Και οι εξαρτώμενες από την χρόνο:
$$\psi_n(x,t)=\psi_n(x)exp(-iE_nt/\hbar)$$

#### Απαραίτητες ρυθμίσεις - βιβλιοθηκών και πεδίου ορισμού

```python
%matplotlib ipympl
import matplotlib
import matplotlib.pyplot as plt 
import numpy
import numpy.polynomial.hermite as Herm
import math

m=1
w=1
hbar=1

dx = 0.1
x_max = 10
x = numpy.arange(-x_max,x_max,dx)

def norm(dx, f):
    return numpy.dot(numpy.conjugate(f), f)*dx

def energy_n(n):
    return hbar*w*(n+0.5)

def psi_n_x(n, x):
    herm_arg = numpy.sqrt(m*w/hbar)*x
    herm_coeffs = numpy.zeros(n+1)
    herm_coeffs[n] = 1
    prefactor = 1./numpy.sqrt(2.**n * math.factorial(n)) * (m*w/(numpy.pi*hbar))**(0.25)
    psi = prefactor * numpy.exp(- herm_arg**2 / 2) * Herm.hermval(herm_arg, herm_coeffs)
    return psi

n=10
psi = psi_n_x(n,x)
prob = numpy.conjugate(psi)*psi
print("Norm = ", dx*numpy.sum(prob))
plt.plot(x, prob)
plt.title(r'Probability Distribution for $\psi_{10}(x)$')
plt.xlabel('x')
plt.ylabel(r'$|\psi_{10}(x)|^2$')
plt.show()
```
