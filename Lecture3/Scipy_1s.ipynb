## Ολοκλήρωση

### Αριθμητική ολοκλήρωση: τετράγωνο

Αριθμητική ολοκλήρωση συνάρτησης του τύπου:

$\displaystyle \int_a^b f(x) dx$

ονομάζεται *αριθμητικό τετράγωνο*, ή απλά *τετραγωνισμός*. Το SciPy παρέχει μια σειρά από συναρτήσεις για διαφορετικά είδη τεταρτημορίου, για παράδειγμα τις «τετράγωνες», «dblquad» και «tplquad» για μονά, διπλά και τριπλά ολοκληρώματα, αντίστοιχα.



```python
from scipy.integrate import quad, dblquad, tplquad
```

Η συνάρτηση "quad" λαμβάνει έναν μεγάλο αριθμό προαιρετικών ορισμάτων, τα οποία μπορούν να χρησιμοποιηθούν για να ρυθμίσετε τη συμπεριφορά της συνάρτησης (δοκιμάστε το "help(quad)" για λεπτομέρειες).

Η βασική χρήση είναι η εξής:

```python
# ορίστε μια απλή συνάρτηση για το ολοκλήρωμα
def f(x):
    return x
```

```python
x_lower = 0 # το κατώτερο όριο του x
x_upper = 1 # το ανώτατο όριο του x

val, abserr = quad(f, x_lower, x_upper)

print("integral value =", val, ", absolute error =", abserr) 
```

Εάν χρειάζεται να περάσουμε επιπλέον ορίσματα στη συνάρτηση integrand, μπορούμε να χρησιμοποιήσουμε το όρισμα λέξης κλειδιού «args»:

```python
def integrand(x, n):
    """
    Bessel function of first kind and order n. 
    """
    return jn(n, x)


x_lower = 0  # το κατώτερο όριο του x
x_upper = 10 # το ανώτατο όριο του x

val, abserr = quad(integrand, x_lower, x_upper, args=(3,))

print(val, abserr) 
```

Για απλές συναρτήσεις μπορούμε να χρησιμοποιήσουμε μια συνάρτηση λάμδα (συνάρτηση χωρίς όνομα) αντί να ορίσουμε ρητά μια συνάρτηση για το ολοκλήρωμα:

```python
from math import  *
from numpy import *
val, abserr = quad(lambda x: exp(-x ** 2), -Inf, Inf)

print( "numerical  =", val, abserr)

analytical = sqrt(pi)
print ("analytical =", analytical)
```

Η ολοκλήρωση περισσότερων διαστάσεων λειτουργεί με τον ίδιο τρόπο:

```python
def integrand(x, y):
    return exp(-x**2-y**2)

x_lower = 0  
x_upper = 10
y_lower = 0
y_upper = 10

val, abserr = dblquad(integrand, x_lower, x_upper, lambda x : y_lower, lambda x: y_upper)

print (val, abserr) 
```

Εδώ θα χρησιμοποιήσουμε τη συνάρτηση sol_ivp. Για περισσότερες πληροφορίες σχετικά με την κλάση "sol_ivp", δοκιμάστε το "help(solve_ivp)". Κάνει σχεδόν το ίδιο πράγμα με το «odeint», αλλά με αντικειμενοστραφή τρόπο.

```python
from scipy.integrate import solve_ivp
```

Ας δουμε τις παραμέτρους εισαγωγής λίγο πιο αναλυτικά:

#### Παράμετροι εισαγωγής:
- __fun(t, y, *args)__: Η συνάρτηση που ορίζει το σύστημα των διαφορικών εξισώσεων. Πρέπει να επιστρέφει το αποτέλεσμα των παραγώγων.
- __t_span__ = (t0, tf): Το αρχικό και τελικό σημείο της ανεξάρτητης μεταβλητής (συνήθως ο χρόνος t ή η θέση x).
- __y0__: Αρχικές συνθήκες (array)
- __method__: Ο αριθμητικός αλγόριθμος που χρησιμοποιείται. Οι πιο κοινοί είναι:
        'RK45' (Runge-Kutta 4/5, προεπιλογή)
        'RK23' (Runge-Kutta 2/3, χρήσιμο για δύσκολες εξισώσεις)
        'BDF' (backward differentiation, καλό για δύσκαμπτα συστήματα)
        'LSODA' (αυτόματη επιλογή μεταξύ άκαμπτων και μη)
- __t_eval__: Πίνακας τιμών του t στα οποία θα υπολογιστούν οι λύσεις.
- __args__: Πρόσθετα ορίσματα που περνούν στη συνάρτηση fun.

#### Τι επιστρέφει η solve_ivp

Το αποτέλεσμα είναι ένα Bunch object, το οποίο λειτουργεί σαν ένα dictionary και περιλαμβάνει τα εξής βασικά πεδία:

- __t__: Μια λίστα ή numpy.ndarray με τα χρονικά (ή χωρικά) σημεία στα οποία υπολογίστηκε η λύση.
- __y__: Ένας πίνακας με τις τιμές των λύσεων για κάθε σημείο του t. Αν η εξίσωση έχει πολλές μεταβλητές, τότε κάθε γραμμή αντιστοιχεί σε μια μεταβλητή.
- __sol__: Μια συνάρτηση που επιτρέπει την αξιολόγηση της λύσης για οποιοδήποτε ενδιάμεσο σημείο (αν επιλέξεις dense_output=True).
- __success__: Boolean που δείχνει αν η ολοκλήρωση ήταν επιτυχής (True ή False).
- __message__: Μήνυμα που περιγράφει την κατάσταση της ολοκλήρωσης.

Λύση της διαφορικής εξίσωσης:

$$\frac{dy}{dt} = -2y$$

με αρχική συνθήκη  $y(0)=1$

```python
def dydt(t, y):
    return -2 * y

t_span = (0, 5)
y0 = [1]

sol = solve_ivp(dydt, t_span, y0, method='RK45', t_eval=np.linspace(0, 5, 100))

plt.plot(sol.t, sol.y[0], label="y(t)")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Λύση της εξίσωσης dy/dt = -2y")
plt.legend()
plt.show()
```

### Ασκηση

Ένα σωματίδιο μάζας m είναι παγιδευμένο σε ένα μονοδιάστατο άπειρο τετραγωνικό φρέαρ μήκους L.

Η χρονικά ανεξάρτητη εξίσωση Schrödinger δίνεται από τη διαφορική εξίσωση:

$$
\frac{d^2\psi}{dx^2} = -\frac{2m}{\hbar^2} (E - V(x)) \psi
$$
και το δυναμικό:
$$ V(x) = \begin{cases} 
0, & 0 < x < L \\
\infty, & x \leq 0 \text{ ή } x \geq L
\end{cases}$$


Συνοριακές συνθήκες:
$\psi(0) = 0, \quad \psi(L) = 0 $

Ζητούμενα:

1. Δείξτε ότι οι επιτρεπτές τιμές της ενέργειας $E_n$ είναι:

$$E_n = \frac{n^2 \pi^2 \hbar^2}{2mL^2}, \quad n = 1,2,3, \dots$$

2. Χρησιμοποιήστε την solve_ivp για να υπολογίσετε και να σχεδιάσετε τη κυματοσυνάρτηση ψ(x) για την τρίτη διεγερμένη στάθμη (n=4).



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Σταθερές
hbar = 1  
m = 1     
L = 2     # Μπορείς να αλλάξεις το L
n = 4     

# Αναλυτική λύση
def analytical_solution(x, n, L):
    return np.sin(n * np.pi * x / L)


# Διάστημα και σωστές αρχικές συνθήκες
x_span = (0, L)
y0 = [0, n * np.pi / L]  # Ψ(0) = 0, Ψ'(0) = nπ/L


# Ορισμός του συστήματος διαφορικών εξισώσεων
def schrodinger(x, y):
    psi, phi = y  # Το y είναι διάνυσμα [ψ(x), φ(x)]
    
    # Πρώτη εξίσωση: dψ/dx = φ
    dpsi_dx = phi  
    
    # Δεύτερη εξίσωση: dφ/dx = - k^2 ψ
    dphi_dx = - (n**2 * np.pi**2 / L**2) * psi  

    return [dpsi_dx, dphi_dx]  # Επιστροφή και των δύο εξισώσεων



# Επίλυση με solve_ivp
x_values = np.linspace(0, L, 100)
sol = solve_ivp(schrodinger, x_span, y0, t_eval=x_values, method="RK45")

# Υπολογισμός της αναλυτικής λύσης
psi_exact = analytical_solution(x_values, n, L)

# Σχεδίαση των αποτελεσμάτων
plt.plot(sol.t, sol.y[0], label=r"Αριθμητική λύση ($\psi_{n}(x)$)", linestyle="dashed")
plt.plot(x_values, psi_exact, label=r"Αναλυτική λύση ($\sin(n\pi x / L)$)", linestyle="solid")
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.xlabel("x")
plt.ylabel(r"$\psi(x)$")
plt.title(f"Σύγκριση Αριθμητικής και Αναλυτικής Λύσης για n={n} και L={L}")
plt.legend()
plt.grid()
plt.show()
```
